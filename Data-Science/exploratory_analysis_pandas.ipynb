10.	Explica cuáles son las diferencias y similitudes entre el par `melt()/pivot()` y el par `stack()/unstack()`. Si tuvieras que elegir una contraparte para `melt()` entre `stack()/unstack()`, ¿cuál elegirías? ¿Por qué?

**melt y pivot**

Similitudes: Las dos transforman de largo a ancho

Diferencias: Melt convierte las columnas en variables y Pivot Convierte los valores únicos en columnas.

Melt preserva columnas no modificadas y pivot puede agregar datos a las columnas index y columns

**stack y unstack**

Similitudes: Transforman la estructura de los datos en Dataframe y permite la organización de los datos.

Diferencias: Uno apila y el otro desapila, stack es de formato largo y unstack de formato ancho

**Elección**

Si hay un índice jerárquico y se desea cambiar la estructura del índice, sería mejor opción utilizar stack y unstack, en caso de no existir un jerarquía se puede utilizar melt y pivot.
